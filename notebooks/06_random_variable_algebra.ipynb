```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Initial parameters
initial_frequency = 1.0
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(initial_frequency * x)

# Create the plot
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)
(line,) = ax.plot(x, y, lw=2)
ax.set_xlim([0, 2 * np.pi])
ax.set_ylim([-1, 1])

# Add a slider for frequency adjustment
axfreq = plt.axes([0.1, 0.1, 0.8, 0.05], facecolor="lightgoldenrodyellow")
slider = Slider(axfreq, "Frequency", 0.1, 5.0, valinit=initial_frequency)


# Update function
def update(val):
    frequency = slider.val
    line.set_ydata(np.sin(frequency * x))
    fig.canvas.draw_idle()


# Connect the update function to the slider
slider.on_changed(update)

# Display the plot
plt.show()
```
